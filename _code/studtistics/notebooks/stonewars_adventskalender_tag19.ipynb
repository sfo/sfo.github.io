```python
import math
import pandas as pd

df_sets = (
    pd.read_csv("https://cdn.rebrickable.com/media/downloads/sets.csv.gz")
    .assign(set_num=lambda df: df["set_num"].str[:-2])
    .query("`set_num`.str.isnumeric() & `set_num`.str.len() > 3")
)

set_nums = df_sets["set_num"]
```

```python
def is_prime(n):
    n = int(n)
    if n < 2 or (n % 2 == 0 and n > 2):
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))

def divisors(n):
    n = int(n)
    return [i for i in range(1, n) if n % i == 0]
```

```python
for i in set_nums:
    if (
        all([is_prime(i[idx]) for idx in range(3)])
        and is_prime(i[0:2])
        and is_prime(i[1:3])
        and is_prime(i[:3])
        and i[-1] != "0"
        and float(int(i[0]) * int(i[1]) * int(i[2])) % int(i[-1]) == 0
    ):
        print(i)
```

Dies sind also alle Setnummern, welche die erste, mathematische Bedingung erfüllen.

Hierfür bedienen wir uns einer zuvor definierten Funktion, welche uns eine Liste eben jener Teiler erzeugt:

```python
n = 21325
divisors(n)
```

Somit gilt es nun zunächst alle Teiler der Setnummer *21318* zu bestimmen, wofür wir die gleiche Funktion, wie zuvor, benutzen:

 Mit dieser Liste gehen wir weiter:

 > [...] bei dem man nur die Ziffern zu vertauschen brauchte, und schon ergab sich die Setnummer eines Sets, das 2025 EOL geht.

 Wenn das gesuchte Set Ende 2025 vom Markt verschwinden soll, muss es sich um ein aktuelles Set mit 5-stelliger Nummer handeln.
 Diese kann sich also lediglich aus dem Teiler "10659" ergeben.

 Um mögliche Kandidaten zu identifizieren, prüfen wir, welche Sets aus den 5 Ziffern der Zahl *10659* bestehen, indem wir die Ziffern jeweils sortieren und mit der Liste der gesuchten Ziffern vergleichen:

```python
n = sorted("10659")
candidate_sets = [set_num for set_num in set_nums if sorted(set_num) == n]
candidate_sets
```

Um herauszufinden, welches dieser Sets am Ende des Jahres 2025 EOL geht, sortieren wir zunächst die Liste nach der Spalte "year".
Dann können wir beginnend mit dem neuesten Set prüfen, welche hiervon auf der Stonewars [EOL-Liste 2025](https://www.stonewars.de/eol-2025/) stehen:

```python
df_sets.query("`set_num` in @candidate_sets").sort_values("year")
```
