This is about some general stuffe like
- number of sets per year
- size of sets over the years
- number of set per theme

```python
import plotly.express as px
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from putil import self_merge
```

```python
df_themes = pd.read_csv("https://cdn.rebrickable.com/media/downloads/themes.csv.gz")
df_sets = pd.read_csv("https://cdn.rebrickable.com/media/downloads/sets.csv.gz").merge(
    self_merge(
        df_themes.rename({"name": "theme"}, axis="columns"), column_name="theme"
    ).set_index("id"),
    how="left",
    left_on="theme_id",
    right_index=True,
)
```

```python
fig = px.histogram(
    df_sets,
    x="theme",
    title="number of sets per theme",
)
fig.show()
```

```python
fig = px.histogram(
    df_sets,
    x="year",
    title="Number of sets released per year",
)
fig.show()
```

```python
fig = px.line(
    df_sets[["year", "num_parts"]].groupby("year").mean().reset_index(),
    x="year",
    y="num_parts",
    title="Average number of parts per set",
)
fig.show()
```

```python
fig = px.bar(
    df_sets[["theme", "num_parts"]].groupby("theme").mean().reset_index().sort_values("num_parts", ascending=False).head(50),
    x="theme",
    y="num_parts",
    title="Average number of parts per theme",
)
fig.show()
```
